# Dynamic time warping for time series data

The problem with time series is that sometimes, they may have different lengths. They may also be taken at different intervals. Thus, we want to use dynamic time warping to make our data of the same length. 

```python
import numpy as np
from dtw import *
import matplotlib.pyplot as plt
```

```python
idx = np.linspace(0,6.28,num=100)
query = np.sin(idx) + np.random.uniform(size=100)/10.0
plt.plot(idx, query)
```

```python
reference = np.cos(idx)
plt.plot(idx, query, label='Noisy sine')
plt.plot(idx, reference, label='Cos')
plt.legend()
```

```python
alignment = dtw(query, reference)
plt.plot(alignment.index1, alignment.index2)  
```

```python
alignmentOBE = dtw(query[44:88], reference,
                     keep_internals=True,
                     step_pattern=asymmetric,
                     open_end=True,open_begin=True)
alignmentOBE.plot(type="twoway",offset=1)  
```

```python
plt.plot(reference);
plt.plot(alignment.index2,query[alignment.index1])  
```
